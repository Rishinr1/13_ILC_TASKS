```python
import pyarrow.parquet as pq
import pandas as pd

def compare_parquet_files(file_path1, file_path2):
    # Read the Parquet files into DataFrames
    df1 = pq.read_table(file_path1).to_pandas()
    df2 = pq.read_table(file_path2).to_pandas()
    
    # Check if the schemas are the same
    if df1.columns.tolist() != df2.columns.tolist():
        print("Schemas are different.")
        print(f"Schema of {file_path1}: {df1.columns.tolist()}")
        print(f"Schema of {file_path2}: {df2.columns.tolist()}")
        return
    
    # Compare the data
    if df1.equals(df2):
        print("Data in both Parquet files is identical.")
    else:
        print("Data in the Parquet files is different.")
        
        # Concatenate the DataFrames side-by-side
        concatenated_df = pd.concat([df1.add_suffix('_file1'), df2.add_suffix('_file2')], axis=1)
        
        # Add a column for the difference in the "Loss" column
        concatenated_df['Loss_Difference'] = concatenated_df['Loss_file1'] - concatenated_df['Loss_file2']
        
        print("Concatenated DataFrame with Loss Difference:")
        print(concatenated_df)

# Define the file paths
file_path1 = r"D:\RISHIN\Rough\ILC2024_EUWS_PLA_WI_EP_BE_EUR_Losses\EP\Lob\GU\ILC2024_EUWS_PLA_WI_EP_BE_EUR_EP_Lob_GU_1.parquet"
file_path2 = r"D:\RISHIN\13_ILC_TASK1\input\EXPORT\ILC2024_EUWS_PLA_WI_EP_BE_EUR_Losses\ILC2024_EUWS_PLA_WI_EP_BE_EUR_Losses\EP\Lob\GU\ILC2024_EUWS_PLA_WI_EP_BE_EUR_EP_Lob_GU_0.parquet"

# Compare the Parquet files
compare_parquet_files(file_path1, file_path2)
```
