```python
import pandas as pd

def compare_and_concatenate_csv_files(path1, path2):
    # Read the CSV files
    df1 = pd.read_csv(path1)
    df2 = pd.read_csv(path2)

    # Check if the dataframes have the same shape
    if df1.shape != df2.shape:
        print("The CSV files have different shapes.")
        return None, None, None

    # Check if the dataframes have the same columns
    if list(df1.columns) != list(df2.columns):
        print("The CSV files have different columns.")
        return None, None, None

    # Check if the dataframes have the same order of data
    if not df1.equals(df2):
        print("The CSV files have different data or order of data.")
        return None, None, None

    # Add a column to show the difference between 'Loss' and 'Sum_Loss' columns
    df1['Loss_Sum_Loss_Difference'] = df1['Loss'] - df1['Sum_Loss']
    df2['Loss_Sum_Loss_Difference'] = df2['Loss'] - df2['Sum_Loss']

    # Concatenate the dataframes
    df3 = pd.concat([df1, df2], ignore_index=True)

    # Add a column to show the difference in 'Loss' between the two dataframes
    df3['Loss_Difference'] = df3['Loss'].diff().fillna(0)

    return df1, df2, df3

# Example usage
df1, df2, df3 = compare_and_concatenate_csv_files('path_to_csv1.csv', 'path_to_csv2.csv')

if df1 is not None and df2 is not None and df3 is not None:
    print("Dataframes are identical and the difference column has been added. Concatenated dataframe created.")
else:
    print("Dataframes are not identical or have different shapes/columns.")
```
